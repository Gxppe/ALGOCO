```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ruta_cub_trad_mnp = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Multiplicacion/Simple/Rectangular_simple_time.txt'
ruta_cub_trad_nxn = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Multiplicacion/Simple/Cuadradas_2_simple_time.txt'
```

```python
ruta_cub_it_mnp = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Multiplicacion/Optimizado/Rectangular_optimizado_time.txt'
ruta_cub_it_nxn = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Multiplicacion/Optimizado/Cuadradas_2_optimizado_time.txt'
```

```python
ruta_stra_nn = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Multiplicacion/Strassens/Cuadradas_2_strassens_time.txt'
```

```python
df_cub_trad_mnp = pd.read_csv(ruta_cub_trad_mnp, sep=' ', header=None, names=['Tamaño', 'Tiempo'])
df_cub_trad_nxn = pd.read_csv(ruta_cub_trad_nxn, sep=' ', header=None, names=['Tamaño', 'Tiempo'])

df_cub_trad_mnp['Tamaño'] = df_cub_trad_mnp['Tamaño'].astype(int)
df_cub_trad_nxn['Tamaño'] = df_cub_trad_nxn['Tamaño'].astype(int)
df_cub_trad_mnp['Tiempo'] = df_cub_trad_mnp['Tiempo'].astype(float)
df_cub_trad_nxn['Tiempo'] = df_cub_trad_nxn['Tiempo'].astype(float)

df_cub_trad_mnp.sort_values(by='Tamaño', inplace=True)
df_cub_trad_nxn.sort_values(by='Tamaño', inplace=True)
```

```python
df_cub_it_mnp = pd.read_csv(ruta_cub_it_mnp, sep=' ', header=None, names=['Tamaño', 'Tiempo'])
df_cub_it_nxn = pd.read_csv(ruta_cub_it_nxn, sep=' ', header=None, names=['Tamaño', 'Tiempo'])

df_cub_it_mnp['Tamaño'] = df_cub_it_mnp['Tamaño'].astype(int)
df_cub_it_nxn['Tamaño'] = df_cub_it_nxn['Tamaño'].astype(int)
df_cub_it_mnp['Tiempo'] = df_cub_it_mnp['Tiempo'].astype(float)
df_cub_it_nxn['Tiempo'] = df_cub_it_nxn['Tiempo'].astype(float)

df_cub_it_mnp.sort_values(by='Tamaño', inplace=True)
df_cub_it_nxn.sort_values(by='Tamaño', inplace=True)
```

```python
df_stra_nn = pd.read_csv(ruta_stra_nn, sep=' ', header=None, names=['Tamaño', 'Tiempo'])
df_stra_nn['Tamaño'] = df_stra_nn['Tamaño'].astype(int)
df_stra_nn['Tiempo'] = df_stra_nn['Tiempo'].astype(float)
df_stra_nn.sort_values(by='Tamaño', inplace=True)
```

```python
plt.figure(figsize=(15, 9))
plt.scatter(df_cub_trad_mnp['Tamaño'], df_cub_trad_mnp['Tiempo'], label='Cúbico Tradicional MxNxP', color='green', alpha=0.5)
plt.scatter(df_cub_it_mnp['Tamaño'], df_cub_it_mnp['Tiempo'], label='Cúbico Optimizado MxNxP', color='red', alpha=0.5)
plt.xlabel('Volumen matriz')
plt.ylabel('Tiempo de ejecución [ms]')
plt.title('Comparación de algoritmos')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.scatter(df_cub_trad_nxn['Tamaño'], df_cub_trad_nxn['Tiempo'], label='Cúbico Tradicional NXN', color='deeppink', alpha=0.5)
plt.scatter(df_cub_it_nxn['Tamaño'], df_cub_it_nxn['Tiempo'], label='Cúbico Iterativo NXN', color='dodgerblue', alpha=0.5)
plt.scatter(df_stra_nn['Tamaño'], df_stra_nn['Tiempo'], label='Strassen NXN', color='lime', alpha=0.5)
plt.xlabel('Tamaño de la matriz ')
plt.ylabel('Tiempo de ejecución (ns)')
plt.title('Comparación de algoritmos')
plt.grid()
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
