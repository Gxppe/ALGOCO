```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

# Rutas de los archivos
ruta_ins_random = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Insertion Time/Random_sorted_insertion_time.txt'
ruta_ins_30 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Insertion Time/Middle_sorted_insertion_time_30.txt'
ruta_ins_70 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Insertion Time/Middle_sorted_insertion_time_70.txt'

# Funcin para cargar datos y convertir tipos
def cargar_y_preparar_datos(ruta, columnas=['Tamao', 'Tiempo']):

        df = pd.read_csv(ruta, sep=" ", header=None, names=columnas)
        df['Tamao'] = df['Tamao'].astype(int)
        df['Tiempo'] = df['Tiempo'].astype(float)  # Usar np.float64 para compatibilidad
        df.sort_values(by=['Tamao'], inplace=True)
        return df


# Cargar y preparar los DataFrames
df_ins_al = cargar_y_preparar_datos(ruta_ins_random)
df_ins_inv = cargar_y_preparar_datos(ruta_ins_30)
df_ins_ord = cargar_y_preparar_datos(ruta_ins_70)

ruta_merge_random = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Merge Time/Random_sorted_merge_time.txt'
ruta_merge_30 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Merge Time/Middle_sorted_merge_time_30.txt'
ruta_merge_70 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Merge Time/Middle_sorted_merge_time_70.txt'

df_merge_al = cargar_y_preparar_datos(ruta_merge_random)
df_merge_inv = cargar_y_preparar_datos(ruta_merge_30)
df_merge_ord = cargar_y_preparar_datos(ruta_merge_70)

ruta_quick_random = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Quick Time/Middle_sorted_quick_time_30.txt'
ruta_quick_30 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Quick Time/Middle_sorted_quick_time_30.txt'
ruta_quick_70 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Quick Time/Middle_sorted_quick_time_70.txt'

df_quick_al = cargar_y_preparar_datos(ruta_quick_random)
df_quick_inv = cargar_y_preparar_datos(ruta_quick_30)
df_quick_ord = cargar_y_preparar_datos(ruta_quick_70)

ruta_introsort_random = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Tim Time/Random_sorted_tim_time.txt'
ruta_introsort_30 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Tim Time/Middle_sorted_tim_time_30.txt'
ruta_introsort_70 = '/home/gxuseppe/Codigo/ALGOCO/Tiempos Registrados/Ordenamiento/Tim Time/Middle_sorted_tim_time_70.txt'

df_introsort_al = cargar_y_preparar_datos(ruta_introsort_random)
df_introsort_inv = cargar_y_preparar_datos(ruta_introsort_30)
df_introsort_ord = cargar_y_preparar_datos(ruta_introsort_70)

# Grficos
plt.figure(figsize=(15, 9))
plt.scatter(df_ins_al['Tamao'], df_ins_al['Tiempo'], label='Insertion sort (arreglo aleatorio)', color='green', alpha=0.5)
plt.scatter(df_merge_al['Tamao'], df_merge_al['Tiempo'], label='Merge sort (arreglo aleatorio)', color='dodgerblue', alpha=0.5)
plt.scatter(df_quick_al['Tamao'], df_quick_al['Tiempo'], label='Quick sort (arreglo aleatorio)', color='red', alpha=0.5)
plt.scatter(df_introsort_al['Tamao'], df_introsort_al['Tiempo'], label='Tim sort (arreglo aleatorio)', color='cyan', alpha=0.5)
plt.xlabel('Tamao del arreglo')
plt.ylabel('Tiempo [ms]')
plt.title('Comparacin de algoritmos')
plt.grid()
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
